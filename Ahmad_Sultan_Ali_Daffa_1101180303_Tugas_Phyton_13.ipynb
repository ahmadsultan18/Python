```
import numpy as np
import matplotlib.pyplot as plt
```

```
def cos(frek,tipe,ht,hr,d):
  #frek = frekuensi (MHz)
  #tipe = urban, suburban, open
  #ht = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
f=1800 #MHz
f1=900
f2=700
loc='urban'
loc1='suburban'
loc2='open'
tower=80 #m
user=1.5 #m
dmin=200 #m
dmax=700 #m
N=200 #titik
d=np.random.randint(dmin,dmax,N)
```

```
print(d)
```

```
d_sort=d[np.argsort(d)]
```

```
def nilaialpha(f,tower,loc,user,d):
  LpUrban2=cos(f,loc,tower,user,d)
  
  Urban_sort2=LpUrban2[np.argsort(d)]
  
  return Urban_sort2
s2=nilaialpha(f,tower,loc,user,d)

print(s2)
```

```
def nilaialpha(f1,tower,loc,user,d):
  LpUrban1=cos(f1,loc,tower,user,d)
  
  Urban_sort1=LpUrban1[np.argsort(d)]
  
  return Urban_sort1
s1=nilaialpha(f1,tower,loc,user,d)

print(s1)
```

```
def nilaialpha(f2,tower,loc,user,d):
  LpUrban=cos(f2,loc,tower,user,d)
  
  Urban_sort=LpUrban[np.argsort(d)]
  
  return Urban_sort
s=nilaialpha(f2,tower,loc,user,d)

print(s)
```

```
plt.figure()
plt.plot(d_sort,s, label='700')
plt.plot(d_sort,s1, label='900')
plt.plot(d_sort,s2, label='1800')
plt.legend()
plt.title('Model Urban')
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```
